```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.regularizers import l2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import MeanSquaredError
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("data_frescoGuard.csv")
```

```python
df.head()
```

```python
X = df.iloc[:, :-1]  # Select all columns except the last one
Y = df.iloc[:, -1] 

X = X.to_numpy()
Y = Y.to_numpy()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=42)
```

```python
print(Y_train.shape)
```

```python
 # Convert DataFrame to NumPy array
X_train = tf.reshape(X_train, shape=(tf.shape(X_train)[0], -1))
print(X_train.shape)
```

```python
model = Sequential(
    [
        Dense(7, activation='relu', name = 'layer1', kernel_regularizer = l2(0.01)),
        Dense(11, activation='relu', name = 'layer2', kernel_regularizer = l2(0.01)),
        Dense(17, activation='relu', name = 'layer3', kernel_regularizer = l2(0.01)),
        Dense(21, activation='relu', name = 'layer4', kernel_regularizer = l2(0.01)),
        Dense(15, activation='relu', name = 'layer5', kernel_regularizer = l2(0.01)),
        Dense(8, activation='relu', name = 'layer6', kernel_regularizer = l2(0.01)),
        Dense(5, activation='relu', name = 'layer7', kernel_regularizer = l2(0.01)),
        Dense(2, activation='relu', name = 'layer8', kernel_regularizer = l2(0.01)),
        Dense(1, activation='relu', name = 'layer9', kernel_regularizer = l2(0.01))
    ]
)
```

```python
model.compile(
    loss = tf.keras.losses.MeanSquaredError(),
    optimizer = tf.keras.optimizers.legacy.Adam(learning_rate=0.01),
)
```

```python
model.fit(X_train, Y_train, epochs=500)
```

```python
X_test = tf.reshape(X_test, shape=(tf.shape(X_test)[0], -1))
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
r2 = r2_score(Y_test, y_pred)
mse = mean_squared_error(Y_test, y_pred)
```

```python
print("MSE = ", mse)
print("R2 = ", r2)
```
